```python
from sklearn.datasets import load_iris
iris = load_iris()
print(iris.data)
#print(iris.DESC)
```

```python
print("Target Classes :\n",iris.target)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(iris.data,iris.target,test_size=0.30)
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=5)
clf.fit(xtrain,ytrain)
y_pred = clf.predict(xtest) #making the prediction on the test data
pre_target = [iris.target_names[i] for i in y_pred]
actual_target = [iris.target_names[i] for i in ytest]
```

```python
print("Predicted","\t\t\tActual","\t\tAnswer")
for i in range(0,len(pre_target)):
    print(i,":",pre_target[i],"\t\t", actual_target[i], "\t\t", end = '\t')
    print('Yes') if pre_target[i]==actual_target[i] else print('No')
```

```python
from sklearn import metrics
print("Confusion Matrix : ",metrics.confusion_matrix(ytest,y_pred))
pri
```
